# Lets train the agent to land on moon correctly

## Observation and action space of the environment

```python
# Lets train the agent to land on moon correctly

# Observation space of the evironment
env = gym.make('LunarLander-v2')
env.reset()
print('Observation space shape', env.observation_space.shape)
print('Sample Observation selected at random', env.observation_space.sample())

# Action space of the environment
print('Action Space shape', env.action_space.n)
print('Action space sample selected at random',  env.action_space.sample())
```

## Vectorized Environment


```python
# We create a vectorized environment
#(method for stacking multiple independent environments into a single environment) of 16 environments,
#this way, we’ll have more diverse experiences during the training.
env = make_vec_env('LunarLander-v2', n_envs = 16)
```

## Create the Model

```python
import gym
env = gym.make('LunarLander-v2')

# Instantiate the agen
model = PPO('MlpPolicy', env, verbose = 1)

# Train the agent
model.learn(total_timesteps = int(2e5))
```

```python
# Define a PPO MlpPolicy architecture
# We use MultiLayerPerceptron (MLPPolicy) because the input is a vector,
# if we had frames as input we would use CnnPolicy

model = PPO(
    policy = 'MlpPolicy',
    env = env,
    n_steps = 1024,
    n_epochs = 4,
    gamma = 0.999,
    gae_lambda = 0.98,
    ent_coef = 0.01,
    verbose =1
)

# Train it for 1,000,000 timesteps
model.learn(total_timesteps=1000000)

# Save the the model
model_name = 'ppo-LunaraLander-v2'
model.save(model_name)
```

```python
# Create a new environment for evaluation
eval_env = gym.make('LunarLander-v2')

mean_reward, std_reward = evaluate_policy(model = model, env = eval_env, n_eval_episodes = 10, render=False, deterministic=True)
print(f"Mean Reward = {mean_reward:.2f} +/- {std_reward}")
```
